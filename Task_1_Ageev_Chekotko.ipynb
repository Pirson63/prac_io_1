Ноутбук выполнил Агеев Иван, 312 группа

```python
import numpy as np
from scipy.optimize import linprog
import matplotlib.pyplot as plt


def nash_equilibrium(A):
    p = []
    q = []
    V = 0
    height = A.shape[0]
    width = A.shape[1]

    c_1 = np.ones(height)
    b_ub_1 = -np.ones(width)
    A_ub_1 = -A.T

    c_2 = -np.ones(width)
    b_ub_2 = np.ones(height)
    A_ub_2 = A

    p = linprog(c_1, A_ub=A_ub_1, b_ub=b_ub_1).x
    V = 1 / np.sum(p)
    q = linprog(c_2, A_ub=A_ub_2, b_ub=b_ub_2).x
    p *= V
    q *= V

    return p, q, V
```

```python
import matplotlib.pyplot as plt

def plot_probability_vector(prob_vector, s):

    x = range(1, len(prob_vector) + 1)

    plt.bar(x, prob_vector, color='skyblue', edgecolor='black')

    plt.xlabel('Номер элемента')
    plt.ylabel('Вероятность')
    plt.title(s)

    plt.ylim(0, 1)

    plt.grid(axis='y', linestyle='--', alpha=0.7)

    plt.show()
```

```python
## TEST

test1 = np.array([[4, 0, 6, 2, 2, 1], [3, 8, 4, 10, 4, 4], [1, 2, 6, 5, 0, 0], [6, 6, 4, 4, 10, 3], [10, 4, 6, 4, 0, 9], [10, 7, 0, 7, 9, 8]])
p, q, V = nash_equilibrium(test1)

print(f"p = {p}")
print(f"q = {q}")

plot_probability_vector(p, 'Стратегия победителя')

plot_probability_vector(q, 'Стратегия проигравшего')
```

```python
#Спектр из одной точки
test_2= np.array([
    [4, 5, 3],
    [7, 6, 8],
    [5, 2, 4]
])

p, q, V = nash_equilibrium(test_2)

print(f"p = {p}")
print(f"q = {q}")

plot_probability_vector(p, 'Стратегия победителя')

plot_probability_vector(q, 'Стратегия проигравшего')
```
